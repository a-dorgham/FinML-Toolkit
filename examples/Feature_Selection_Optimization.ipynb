```python
from ml_toolkit.model_executor import ModelExecutor
from ml_toolkit.visualizer import Visualizer
```

```python
from itertools import combinations
# Define all possible feature subsets
all_features = ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'MACD']
feature_subsets = sum([list(combinations(all_features, r)) for r in range(1, len(all_features) + 1)], [])
feature_subsets = [list(subset) for subset in feature_subsets]
```

```python
# Define all possible feature subsets
feature_subsets = [
    ('Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'MACD'),
    ('Open', 'High', 'Low', 'Close', 'Volume'),
    ('RSI', 'MACD', 'Volume'),
    ('RSI', 'MACD', '%K', '%D'),
    ('%K', '%D'),
    ('RSI', 'MACD', 'Close')
]

# Initialize a dictionary to store results
results = {}

# Define the objective function
def objective_manual(features):
    Visualizer.print_hline_with_background(f"Optimization: feature {features}", font_size="20px", color="#2464b3", background_color="#daeb94")
    
    # Call the model execution with fixed hyperparameters
    model, mse, _, _, _, _, _, _, _ = ModelExecutor.execute_model(
        file_path='../fin_data/GBP_USD_H4.pkl',
        start_date="2022-1-1",
        end_date="2024-07-29 21:00",
        features = features,
        imbalance=True,
        imb_type='smote_enn',
        model_type='Conv1D+Attention', 
        model_style='Functional', 
        model_layers=3, 
        units=64,  
        dropout_rate=0.2, 
        learning_rate=0.001,
        epochs=10, 
        batch_size=64  
    )
    print(f"MSE for features {features}: {mse}")
    return mse

# Loop through each feature subset and evaluate
for subset in feature_subsets:
    mse = objective_manual(subset)
    results[subset] = mse  # Store the result

# Print the results
Visualizer.print_hline_with_background(f"Results summary:", font_size="20px", color="#2464b3", background_color="#daeb94")

for features, mse in results.items():
    print(f"Features: {features}, MSE: {mse}")

```

```python
import pandas as pd
import plotly.graph_objects as go
from ml_toolkit.data_handler import DataHandler
from ml_toolkit.feature_engineer import FeatureEngineer
from ml_toolkit.visualizer import Visualizer
```

```python
file='../fin_data/GBP_USD_H4.pkl'
start_date="2024-01-12 07:30"
end_date="2024-07-29 21:00"

# Load data
df_raw = pd.read_pickle(file)
df = DataHandler.load_data(df_raw, start_date, end_date)

# Plot data with peaks
fig = go.Figure()
fig = Visualizer.plot_with_peaks(df, fig=fig, plot_data=True, plot_ma=False, plot_peaks=False)

# Plot peaks
end_date="2024-07-26 21:00"

# Load the data
df = DataHandler.load_data(df_raw, start_date, end_date)
# Define features
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'MACD', '%K', 'CO', 'CU']
            
# Feature Engineering
df, time, X, y = FeatureEngineer.add_features(df=df, features=features)
df.drop(['Signal'], axis=1, inplace=True)
df['Signal'] = FeatureEngineer.find_peaks_and_valleys(df=df[['Time', 'Close']], 
                                      distance=5, prominence=0.01)

# Identify peaks/valleys
peak_indices = (df['Signal'] > 1.9).sum()
valley_indices = (df['Signal'] < 0.1).sum()

# Plot data with peaks
Visualizer.plot_with_peaks(df, fig=fig, plot_data=False, plot_ma=True, plot_peaks=True, plot_stoch=True)
fig.show()
```

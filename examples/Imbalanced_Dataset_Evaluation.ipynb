```python
from ml_toolkit.model_executor import ModelExecutor
from ml_toolkit.visualizer import Visualizer
```

```python
# Define all possible feature subsets: 
# 'smote', 'adasyn', 'random_oversampler', 'random_undersampler', 'tomek_links', 'nearmiss', 'smote_tomek', 'smote_enn'
imb_models_subsets = [ 'smote_tomek', 'smote_enn'
                   ]

# Initialize a dictionary to store results
results = {}

# Define the objective function
def objective_manual(imb_model):
    Visualizer.print_hline_with_background(f"Optimization: method {imb_model}", font_size="20px", color="#2464b3", background_color="#daeb94")
    
    # Call the model execution with fixed hyperparameters
    model, mse, _, _, _, _, _, _, _ = ModelExecutor.execute_model(
        file_path='../fin_data/GBP_USD_H4.pkl',
        start_date="2022-1-1",
        end_date="2024-07-29 21:00",
        features = ['Close', 'Volume', 'RSI', 'MACD'],
        imbalance=True,
        imb_type=imb_model,
        model_type='Conv1D+Attention', 
        model_style='Functional',
        model_layers=3,
        units=64, 
        dropout_rate=0.2, 
        learning_rate=0.001, 
        epochs=10, 
        batch_size=64 
    )
    print(f"MSE for imb_model {imb_model}: {mse}")
    return mse

# Loop through each feature subset and evaluate
for subset in imb_models_subsets:
    mse = objective_manual(subset)
    results[subset] = mse  # Store the result

# Print the results
Visualizer.print_hline_with_background(f"Results summary:", font_size="20px", color="#2464b3", background_color="#daeb94")
for imb_model, mse in results.items():
    print(f"Features: {imb_model}, MSE: {mse}")

```
